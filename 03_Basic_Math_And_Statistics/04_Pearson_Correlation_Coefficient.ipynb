### Segment 5 - Starting with parametric methods in pandas and scipy

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sb
from pylab import rcParams

import scipy
from scipy.stats.stats import pearsonr
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 8,6
sb.set_style('whitegrid')
```

### The Pearson Correlation

```python
address = '../Datasets/mtcars.csv'

cars = pd.read_csv(address)
cars.columns = ['car_names', 'mpg', 'cyl', 'disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']
```

```python
sb.pairplot(cars)
```

```python
x = cars[['mpg','hp','qsec','wt']]
sb.pairplot(x)
```

```python
mpg = cars.mpg
hp = cars.hp
qsec = cars.qsec
wt = cars.wt

pearsonr_coefficient, pvalue = pearsonr(mpg, hp)
print(f"Pearson Correlation Coefficient for \"mpg\" and \"hp\": {pearsonr_coefficient}")
```

```python
pearsonr_coefficient, pvalue = pearsonr(mpg, qsec)
print(f"Pearson Correlation Coefficient for \"mpg\" and \"qsec\": {pearsonr_coefficient}")
```

```python
pearsonr_coefficient, pvalue = pearsonr(mpg, wt)
print(f"Pearson Correlation Coefficient for \"mpg\" and \"wt\": {pearsonr_coefficient}")
```
